ابتدا نموداری برای مقایسه اهمیت روزهای هفته میکشیم و چون این ویژگی در ۷ ویژگی جدا پخش شده باید آنها را یکی کنیم.  به نظر میرسد در اواسط هفته احتمال آنکه خبر زیاد به اشتراک گذاشته شود کمتر است و این میزان در روزهای پایانی هفته به بیشترین میزان خود میرسد. هر چند مجموع کل اخبار در این روزها کمتر است.

```python
weekday = onlion_news_df.columns.values[31:38]
dataChannel =  onlion_news_df.columns.values[13:19]
dataChannel
```

```python
Unpopular=onlion_news_df[onlion_news_df[' shares']<1400]
Popular =onlion_news_df[onlion_news_df[' shares']>=1400]
Unpopular_day = Unpopular[weekday].sum().values
Popular_day = Popular[weekday].sum().values

fig = plt.figure(figsize = (13,5))
plt.title(" popular/unpopular news over different day of week", fontsize = 16)
plt.bar(np.arange(len(weekday)), Popular_day, width = 0.3, align="center", color = 'r', \
          label = "popular")
plt.bar(np.arange(len(weekday)) - 0.3, Unpopular_day, width = 0.3, align = "center", color = 'b', \
          label = "unpopular")
plt.xticks(np.arange(len(weekday)),weekday)
plt.ylabel("Count", fontsize = 12)
plt.xlabel("Days of week", fontsize = 12)
    
plt.legend(loc = 'upper right')
plt.tight_layout()
plt.show()
```

نمودار دوم برای مقایسه اهمیت موضوع خبر است.  موضوعات جهان و تفریحات از اشتراک گذاری کمتری برخوردارند و اخبار مربوط به شبکه های اجتماعی و تکنولوژی از بیشترین اشتراک گذاری برخوردارند.

```python
Unpopular_DC = Unpopular[dataChannel].sum().values
Popular_DC = Popular[dataChannel].sum().values
fig = plt.figure(figsize = (13,5))
plt.title("popular/unpopular news over different data channel", fontsize = 16)
plt.bar(np.arange(len(dataChannel)), Popular_DC, width = 0.3, align="center", color = 'r', \
          label = "popular")
plt.bar(np.arange(len(dataChannel)) - 0.3, Unpopular_DC, width = 0.3, align = "center", color = 'b', \
          label = "unpopular")
plt.xticks(np.arange(len(dataChannel)), dataChannel)
plt.ylabel("Count", fontsize = 12)
plt.xlabel("Data Channel", fontsize = 12)
    
plt.legend(loc = 'upper right')
plt.tight_layout()
plt.show()
```
